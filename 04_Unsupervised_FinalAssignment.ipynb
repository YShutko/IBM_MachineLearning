```python
from sklearn.pipeline import Pipeline
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier
from collections import Counter
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
```

```python
X = data.drop('Activity', axis=1)
y = data.Activity
print(Counter(y))
```

```python
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42, stratify=y)
print(Counter(y_train))
print(Counter(y_test))
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

## GradientBoostingClassifier

```python
gbc = GradientBoostingClassifier(max_features=4, n_estimators=400, random_state=42,
                           subsample=0.5)
gbc_ = gbc.fit(X_train, y_train)
y_pred_gbc = gbc.predict(X_test)
print(accuracy_score(y_test,y_pred_gbc))
```

```python
def get_avg_score(n):
    pipe = [
       # ( 'scaler', StandardScaler()),
        ('pca', PCA(n_components=n)),
        ('estimator',GradientBoostingClassifier(max_features=4, n_estimators=400, random_state=42,
                           subsample=0.5))
    ]
    pipe = Pipeline(pipe)
    scores = []
    pipe.fit(X_train, y_train)
    scores.append(accuracy_score(y_test, pipe.predict(X_test)))
    return np.mean(scores)


ns = [10, 20, 50, 100, 150, 200, 300, 400]
score_list = [get_avg_score(n) for n in ns]
```

```python
score_list
```

```python
sns.set_context('talk')

ax = plt.axes()
ax.plot(ns, score_list)
ax.set(xlabel='Number of Dimensions',
       ylabel='Average Accuracy',
       title=' Gradient Boosting Classifier Accuracy vs Number of dimensions on the Human Activity Dataset')
ax.grid(True)
```

```python

pipe = [
       # ('scaler', StandardScaler()),
        ('pca', PCA(n_components=100)),
        ('estimator',GradientBoostingClassifier(max_features=4, n_estimators=400, random_state=42,
                           subsample=0.5))
    ]
pipe = Pipeline(pipe)
pipe.fit(X_train, y_train)
accuracy_score(y_test, pipe.predict(X_test))
```

## Logistic Regression

```python
lgcv = LogisticRegression(C=0.1, random_state=42, max_iter=10000, fit_intercept=True,solver='liblinear', penalty = "l2")
lgcv_= lgcv.fit(X_train, y_train)
y_pred_lgcv = lgcv_.predict(X_test)
print(accuracy_score(y_test,y_pred_lgcv))
```

```python
def get_avg_score(n):
    pipe = [
       # ('scaler', StandardScaler()),
        ('pca', PCA(n_components=n)),
        ('estimator',LogisticRegression(C=0.1, random_state=42, max_iter=10000, fit_intercept=True,solver='liblinear'))
    ]
    pipe = Pipeline(pipe)
    scores = []
    pipe.fit(X_train, y_train)
    scores.append(accuracy_score(y_test, pipe.predict(X_test)))
    return np.mean(scores)


ns = [10, 20, 50, 100, 150, 200, 300, 400]
score_list = [get_avg_score(n) for n in ns] 
```

```python
pipe = [
        ('scaler', StandardScaler()),
        ('pca', PCA(n_components=100)),
        ('estimator',LogisticRegression(C=0.1, random_state=42, max_iter=10000, fit_intercept=True,solver='liblinear'))
    ]
pipe = Pipeline(pipe)
pipe.fit(X_train, y_train)
accuracy_score(y_test, pipe.predict(X_test))
```
